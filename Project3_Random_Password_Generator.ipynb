# Project 3 - Random Password Generator

```
import random
import string

print("="*50)
print("SECURE RANDOM PASSWORD GENERATOR")
print("="*50)
length=int(input("Enter password length (minimum 8): "))
if length<8:
    print("Password length should be at least 8.")
else:
    chars=[random.choice(string.ascii_uppercase),random.choice(string.ascii_lowercase),random.choice(string.digits),random.choice(string.punctuation)]
    chars+=random.choices(string.ascii_letters+string.digits+string.punctuation,k=length-4)
    random.shuffle(chars)
    print("Generated Password:","".join(chars))
```
